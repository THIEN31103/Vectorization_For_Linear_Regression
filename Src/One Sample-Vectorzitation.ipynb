```python
import numpy as np
```

```python
x = np.array([1, 6.7])
theta = np.array([0.049, -0.34])
x, theta
```

```python
#compute predict
def predict(x, theta):
    return x.T.dot(theta)

#compute loss
def compute_loss(y_hat, y):
    return (y_hat - y)*(y_hat - y)

#compute gradient
def compute_gradient(y_hat, y, X):
    return 2*X*(y_hat - y)

#update weights
def update_weights(theta, gradient, lr):
    return theta - gradient*lr
```

```python
#test
y = 9.1
lr = 0.01
y_hat = predict(x, theta)
gradient = compute_gradient(y_hat, y, x)
theta = update_weights(theta, gradient, lr)

theta
```

```python
# data
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt(r'E:\AIO\Vectorization_For_Linear_Regression\Data\data.csv', delimiter=',')
N = len(data)

areas = data[:, 0].reshape(1,N)
y_label = data[:, 1].reshape(N,)

X = np.vstack([np.ones((1,N)), areas])
print(X)
print(y)
```

```python
theta = np.array([0.049, -0.34])
```

```python
lr = 0.01
num_epoch = 10
```

```python
losses = []
for epoch in range(num_epoch):

    for i in range(N):

        #get a sample
        x = X[:,i]
        y = y_label[i]

        #compute predict
        y_hat = predict(x, theta)

        #compte loss
        loss = compute_loss(y_hat, y)
        losses.append(loss)

        #compute gradient
        gradient = compute_gradient(y_hat, y, x)

        #update weights
        theta = update_weights(theta, gradient, lr)

print(theta)
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
```

```python
x_data = range(2, 8)

predict_b, predict_w = theta
y_data = [x*predict_w + predict_b for x in x_data]
plt.plot(x_data, y_data, 'r')

plt.scatter(areas, y_label)
plt.xlim(3, 7)
plt.ylim(4, 10)
plt.show()
```
