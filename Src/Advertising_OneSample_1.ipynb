```python
# dataset
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt

data = genfromtxt(r'E:\AIO\Vectorization_For_Linear_Regression\Data\advertising.csv', delimiter=',', skip_header=1)

N = data.shape[0]
X = data[:,:3]
y = data[:,3:]

X_b = np.hstack((np.ones((N, 1)), X))
```

```python
print(X_b.shape)
print(X_b[:5])
```

```python
print(y.shape)
print(y[:5])
```

```python
n_epochs = 50
learning_rate = 0.00001

# khởi tạo giá trị tham số
thetas = np.random.randn(4, 1)

thetas_path = [thetas]
losses = []

for epoch in range(n_epochs):
    for i in range(N):
        # lấy 1 sample            
        xi = X_b[i]        
        xi = np.reshape(xi, (4, 1))
        yi = y[i:i+1]

        # tính output 
        y_hat = xi.T.dot(thetas)

        # tính loss li
        li = (y_hat - yi)*(y_hat - yi)

        # tính gradient cho loss
        g_li = 2*(y_hat - yi)

        # tính gradient 
        gradients = xi*g_li

        # update giá trị theta
        thetas = thetas - learning_rate*gradients

        # logging
        thetas_path.append(thetas)            
        losses.append(li[0][0])
```

```python
import matplotlib.pyplot as plt
# in loss cho 500 sample đầu
x_axis = list(range(500))
plt.plot(x_axis,losses[:500], color="r")
plt.show()
```

```python
import numpy as np

data = np.genfromtxt(r'E:\AIO\Vectorization_For_Linear_Regression\Data\advertising.csv', 
                     delimiter=',', 
                     skip_header=1)
print(data.shape)
```

```python
N = 150
n_epochs = 50
lr = 0.00001

# khởi tạo giá trị tham số
thetas = np.random.randn(4,)
losses = []

for epoch in range(n_epochs):
    for i in range(N):
        # get 1 sample          
        sample_i = data[i]
        
        # convert to vector
        x_1 = sample_i[0]
        x_2 = sample_i[1]
        x_3 = sample_i[2]
        y_i = sample_i[3]
        vector_x = np.array([1, x_1, x_2, x_3])
        
        # compute output 
        y_hat = vector_x.T.dot(thetas)

        # compute loss li
        loss = (y_hat - y_i)*(y_hat - y_i)

        # compute gradient 
        gradients = 2*(y_hat - y_i)*vector_x

        # update 
        thetas = thetas - lr*gradients

        # logging          
        losses.append(loss)
```

```python
import matplotlib.pyplot as plt

# in loss cho 500 sample đầu
x_axis = list(range(500))
plt.plot(x_axis,losses[:500], color="r")
plt.show()
```
